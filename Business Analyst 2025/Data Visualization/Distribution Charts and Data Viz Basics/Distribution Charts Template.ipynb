```python
# Libraries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

# Histogram

Histograms are best suited for continuous or quantitative data that represents measurements or values within ranges, as they effectively visualize the distribution and frequency of data points across specified intervals.


```python
# Defining the plot size
fig, ax = plt.subplots(figsize = (15, 6))

# Histogram
ax.hist(df.Total,
        bins=20,                    # Number of bins
        edgecolor='black',          # Color of the bin edges
        color='red',                # Color of the bars
        alpha=0.5,                  # Transparency of the bars
        # range=(400, 600)          # Optional range for the histogram
        )

# Add a Vertical line
ax.axvline(df.Total.median(),
           color = 'black',
           ls = "dashed",
           label = "Median Total Power")

# Add a horizontal line
ax.axhline(50,
           color = 'blue',
           ls = 'dotted',
           label = "Line at frequency of 50")


# Plotting
ax.legend() # Display legend
plt.show()
```

# Box Plot - Analyzing the Attack stat by generation

Boxplots are ideal for visualizing the distribution of continuous data by displaying its median, quartiles, and potential outliers, allowing you to easily see the spread and skewness of the data.

```python
# Defining the plot size
fig, ax = plt.subplots(figsize = (15, 6))
sns.set_theme(style = "darkgrid")

# Set the Order
order = df.groupby(
    by = ["Generation"])["Attack"].median().sort_values(
        ascending = False).index
# Box Plot
sns.boxplot(x = df.Generation,
            y = df.Attack,
            palette = "Spectral",
            hue = df.Generation,
            legend = False,
            saturation = 2,
            orient = 'v',
            whis = 1.5,
            linewidth = 2,
            width = 0.8,
            flierprops = {"marker": "*"},
            fliersize = 10,
            order =  order)

# # Add a data points to the box plot
# ax = sns.swarmplot(x = df.Generation,
#                    y = df.Attack,
#                    color = "grey",
#                    order = order)

plt.show()
```

# Violin Plot - Total Stats per Generation

Violin plots are excellent for visualizing the distribution of continuous data, as they combine features of boxplots and density plots to show the data's distribution shape, central tendency, and spread, including potential multimodality.

```python
# Defining the plot size
fig, ax = plt.subplots(figsize = (15, 6))
sns.set_theme(style = "whitegrid")

# Set the Order
order = df.groupby(
    by = ["Generation"])["Total"].median().sort_values(
        ascending = False).index

# Violin Plot
sns.violinplot(
    x=df.Generation,
    y=df.Total,
    palette="Pastel1",
    saturation=2,
    hue=df.Legendary,
    orient='v',
    linewidth=2,
    order=order,
    density_norm="width",  # Use density_norm instead of scale
    bw_adjust=0.2  # Use bw_adjust instead of bw
)


plt.show()
```

# Ridge Line Chart - Speed Stat per Generation

Ridge line chart is a way to show the distribution of data across different groups or categories by stacking multiple density plots on top of each other. This lets you see patterns, trends, and differences between groups in one view.

```python
# Group Speed by generation
df_speed = df.groupby(['Generation', 'Speed']).agg({'Speed': 'count'}).rename(
    columns = {'Speed': 'count'}).reset_index()
df_speed.head()
```

```python
# Store and normalize speed and count data for each generation
dict_values = {}
for gen in df.Generation.unique():
    # Store the speed data for the current generation
    dict_values[f'x_{gen}'] = df_speed[df_speed['Generation'] == gen]['Speed']

    # Store the count data for the current generation and normalize it
    dict_values[f'y_{gen}'] = df_speed[df_speed['Generation'] == gen]['count']
    dict_values[f'y_{gen}'] = (
        dict_values[f'y_{gen}'] - dict_values[f'y_{gen}'].min()) \
        / (dict_values[f'y_{gen}'].max() - dict_values[f'y_{gen}'].min())
```

```python
# Start the plot
fig = go.Figure()

for index, gen in enumerate(df.Generation.unique()):
  # Build the chart
  fig.add_trace(go.Scatter(x = [df.Speed.min() -10, df.Speed.max() +10],
                           y = np.full(2, len(df.Generation.unique()) - index),
                           mode = 'lines',
                           line_color = 'white'))

  # Add the data
  fig.add_trace(go.Scatter(x = dict_values[f'x_{gen}'],
                           y = dict_values[f'y_{gen}'] + (
                              len(df.Generation.unique()) - index
                           ) +0.1,
                           fill = 'tonexty',
                           name = f'{gen}'))

  # Add some text to the figure
  fig.add_annotation(x = df.Speed.min(),
                     y = len(df.Generation.unique()) - index,
                     text = f'{gen}',
                     showarrow = False,
                     yshift = 10)

# Customizing the layout
fig.update_layout(title = "Development of Pokemon Speed per Generation",
                  showlegend = False,
                  xaxis = dict(title = 'Speed'),
                  yaxis = dict(title = 'Generation',
                               showticklabels = False))


fig.show()
```
